```python
#replacing zeros with mean of that columns
df['exper']=df['exper'].replace(0,df['exper'].mean())
df['union']=df['union'].replace(0,df['union'].mean())
df['goodhlth']=df['goodhlth'].replace(0,df['goodhlth'].mean())
df['black']=df['black'].replace(0,df['black'].mean())
df['female']=df['female'].replace(0,df['female'].mean())
df['married']=df['married'].replace(0,df['married'].mean())
df['service']=df['service'].replace(0,df['service'].mean())
```

```python
#After replacing zeros with mean of the columns
df[df['exper']==0].shape[0]
df[df['union']==0].shape[0]
df[df['goodhlth']==0].shape[0]
df[df['black']==0].shape[0]
df[df['female']==0].shape[0]
df[df['married']==0].shape[0]
df[df['service']==0].shape[0]
```

```python
# LINE PLOT
x=df['black']
y=df['married']
plt.plot(x,y,marker="o",mec='r',mfc='pink')
plt.xlabel("Black")
plt.ylabel("Married")
plt.title("Comparsion between black and married features of dataset")
plt.show()
```

```python
sns.barplot(x="educ",y="service",data=df,color="black")
plt.show()
```

```python
#Bar plot
x=df['black']
y=df['married']
plt.barh(x,label="wage",width=2,height=2,linestyle="--")
plt.barh(y,label="goodhlth",width=2,height=2,linestyle='-.')
plt.title("comparsion")
plt.legend()
plt.grid()
plt.show()
```

```python
#Histogram of each feature
df.hist(figsize=(10,10),grid=True)
plt.show()
```

```python
#scatter plot
x=df['service']
y=df['educ']
plt.scatter(x,y)
plt.show()
```

```python
#distplot(univariant)
x=df['educ']
sns.distplot(x,label="female",hist=True,kde=True,color='r')
plt.show()
```

```python
#distplot without hist
x=df['educ']
sns.distplot(x,label="female",hist=False,kde=True,color='r')
plt.show()
```

```python
#subplot
#plot1
x=df['service']
y=df['black']
plt.subplot(1,2,1)
plt.plot(x,y)


#plot2
x=df['married']
y=df['educ']
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
#heatmap
#Analyzing relationships between variables
#correlation analysis
import seaborn as sns
corrmat=df.corr()
top_corr_features=corrmat.index
plt.figure(figsize=(10,10))
sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
```
